### 1. Download and explore dataset

Toronto city has 11 boroughs and 103 neighbourhoodsIn order to segement the neighborhoods and explore them, we will essentially need a dataset that contains the 11 boroughs and the neighborhoods that exist in each borough as well as the the latitude and logitude coordinates of each neighborhood. Here is the link to the dataset:https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M

Using web scraping library of python,BeautifulSoup let's access this dataset

```python
import bs4 as bs
import urllib.request

sauce = urllib.request.urlopen("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").read()
soup = bs.BeautifulSoup(sauce, 'lxml')
```

```python
# Writing a function to extract table function from webpage

def get_list(final):
    table_rows = final.find_all('tr')
    table_list = []
    for tr in table_rows:
        td = tr.find_all('td')
        row = [i.text for i in td]
        table_list.append(row)
    return table_list
```

```python
# Calling the function to get the table content

output = get_list(soup.find('table'))
output
```

Converting above to list to pandas dataframe and cleaning it

```python
df = pd.DataFrame(output)
df = df.iloc[1:].reset_index(drop=True)
df.columns = ['PostalCode', 'Borough', 'Neighbourhood']
df['Neighbourhood'] = df['Neighbourhood'].str.replace('\n','')
df.head()
```

```python
# Processing the cells that have not assigned borough
df = df[df['Borough'] != 'Not assigned']
df.head()
```

```python
# Making same postalcode neighbourhoods in single row
df = df.groupby(['PostalCode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
df.head()
```

```python
# Assigning neighbourhood same as its borough if it has 'Not assigned' value
for i in df.index:
    if(df.at[i, 'Neighbourhood'] == 'Not assigned'):
        df.at[i, 'Neighbourhood'] = df.at[i, 'Borough']
df.head()    
```

```python
# Merging both dataframes to get neighbourhoods along with their coordinates
result_df = pd.merge(df, df_geo, on='PostalCode')
result_df.head()
```

## 2. Explore neighbourhoods of North York borough of Toronto city

```python
# Create a function to repeat same process to all neighbourhoods in our dataframe.Get top 100 venues in the radius of 500 meters of all neighbourhoods

LIMIT=100
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    venues_list = []
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
     # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name,
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])
        
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood',                 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# Calling the above function

NY_venues = getNearbyVenues(names=NY_df['Neighbourhood'],
                                   latitudes=NY_df['Latitude'],
                                   longitudes=NY_df['Longitude']
                                  )
```

### Analyse each neighbourhood


```python
# Onehot encoding
NY_onehot = pd.get_dummies(NY_venues[['Venue Category']], prefix='', prefix_sep='')

# add neighborhood and borough column back to dataframe
NY_onehot['Neighbourhood'] = NY_venues['Neighbourhood'] 

NY_onehot
# move neighborhood column to the first column
fixed_columns = [NY_onehot.columns[-1]] + list(NY_onehot.columns[:-1])
NY_onehot = NY_onehot[fixed_columns]

NY_onehot
```

```python
# Let's examine new dataframe size
print(NY_onehot.shape)
```

```python
NY_grouped = NY_onehot.groupby('Neighbourhood').mean().reset_index()
NY_grouped.head()
```

```python
# Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category
# Getting the size of new dataframe
NY_grouped.shape
```

**Printing each neighbourhood along with top 5 most common venues**

```python
num_top_venues = 5
tmp = pd.DataFrame()
for hood in NY_grouped['Neighbourhood']:
    print("----"+hood+"----")
    temp = NY_grouped[NY_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[2:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    temp1 = temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues)
    print(temp1)
    tmp = tmp.append(temp1.head(1))
    print('\n')
print(tmp)    
```

```python
# Now, Let's visualise the 1st most common venues in each neighbourhood
import matplotlib.pyplot as plt
#tmp = tmp.set_index('venue')
tmp.plot(kind='bar', figsize=(8,6))
plt.show()

```

```python
# write a function to sort venues in descending order

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# Now let's create the new dataframe and display the top 10 venues for each neighborhood

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighbourhoods_venues_sorted = pd.DataFrame(columns=columns)
neighbourhoods_venues_sorted['Neighbourhood'] = NY_grouped['Neighbourhood']

for ind in np.arange(NY_grouped.shape[0]):
    neighbourhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(NY_grouped.iloc[ind, :], num_top_venues)

neighbourhoods_venues_sorted.head()
```

### Cluster Neighbourhoods

```python
# set number of clusters
kclusters = 3

NY_grouped_clustering = NY_grouped.drop(columns=['Neighbourhood'])

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(NY_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:30]
```

**Let's create a new dataframe that includes the cluster as well as the top 5 venues for each neighborhood.**

```python
# add clustering labels

#neighbourhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
neighbourhoods_venues_sorted.head()
#NY_merged = NY_d
#NY_df.drop(columns='Borough', inplace=True)
# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
NY_merged = NY_df.join(neighbourhoods_venues_sorted.set_index('Neighbourhood'), on=['Neighbourhood'])

NY_merged.head() # check the last columns!
```

```python
NY_merged = NY_merged[NY_merged['Neighbourhood'] != 'Newtonbrook,Willowdale']
#NY_merged['Cluster Labels'].isnull()
NY_merged
```

**Let's visualize above dataframe**

```python
NY_merged['Cluster Labels'] = NY_merged['Cluster Labels'].astype(int)
NY_merged.head()
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, pos, cluster in zip(NY_merged['Latitude'], NY_merged['Longitude'], NY_merged['Neighbourhood'], NY_merged['Cluster Labels']):
    label = folium.Popup(str(pos) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
